```python
def score_calc(p, t1_stats, t2_stats):
    FG_PCT = p[0]
    FT_PCT = p[1]
    FG3_PCT = p[2]
    DREB = p[3]
    OREB = p[4]
    BLK = p[5]
    AST = p[6]
    t1_score = FG_PCT*t1_stats[0]+FT_PCT*t1_stats[1]+FG3_PCT*t1_stats[2]+DREB*t2_stats[3]+OREB*t1_stats[4]-BLK*t2_stats[5]+AST*t1_stats[6]
    t2_score = FG_PCT*t2_stats[0]+FT_PCT*t2_stats[1]+FG3_PCT*t2_stats[2]+DREB*t1_stats[3]+OREB*t2_stats[4]-BLK*t1_stats[5]+AST*t2_stats[6]
    
    print('Team 1 score:',t1_score)
    print('Team 2 score:',t2_score)
    
```

```python
DTR = [0.407,0.785,0.299,31.7,10.7,4.8,21.3]
OKC = [0.407,0.743,0.302,36.0,10.1,4.7,19.7]

score_calc(results.params,DTR,OKC)
```

```python
PHI = [0.475,0.813,0.357,33.7,8.8,6,22.5]
CHAR = [0.460,0.742,0.386,33.6,10.5,5.4,26.3]

score_calc(results.params,PHI,CHAR)
```
